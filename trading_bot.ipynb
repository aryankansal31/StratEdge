```python
class TradingRequirement:
    def __init__(self, growwapi: GrowwAPI):
        self.__growwapi = growwapi

    @property
    def growwapi(self):
        return self.__growwapi

    @property
    def getStrikePrice(self):
        return "57000"

    def getNearestExpiryDateForIndex(self, index_name: str):
        instrument_df = self.getInstruments()
        nearest_expiry = instrument_df[(instrument_df['underlying_symbol'] == index_name.upper())]['expiry_date'].min()
        return nearest_expiry

    def getInstruments(self):
        instrument_df = self.growwapi.get_all_instruments()
        instrument_df['expiry_date'] = pd.to_datetime(instrument_df['expiry_date'])
        return instrument_df


    def getTradingSymbolFnO(self, index_name: str, type: str):
        instrument_df = self.getInstruments()
        trading_symbol_row = instrument_df[
            (instrument_df["underlying_symbol"] == index_name.upper())
            & (instrument_df["instrument_type"] == type.upper())
            & (instrument_df["strike_price"] == self.getStrikePrice)
            & (instrument_df["expiry_date"] == self.getNearestExpiryDateForIndex(index_name=index_name))
        ]
        if trading_symbol_row.empty:
            return None
        return trading_symbol_row['trading_symbol'].iloc[0]
```

# User Details

```python
class GetUserDetails:
    def __init__(self, growwapi: GrowwAPI):
        self.__growwapi = growwapi
        self.__segment_cash = growwapi.SEGMENT_CASH
        self.__segment_derivatives = growwapi.SEGMENT_FNO

    @property
    def growwapi(self):
        return self.__growwapi

    def getUserHoldings(self):
        user_holdings = self.growwapi.get_holdings_for_user()
        if(not user_holdings['holdings']):
            return "No holdings found"
        return pd.DataFrame(user_holdings['holdings'])

    def getUserPositionsEquity(self):
        user_positions = self.growwapi.get_positions_for_user(segment=self.__segment_cash)
        return user_positions
    
    def getUserPositionsFnO(self):
        user_positions = self.growwapi.get_positions_for_user(segment=self.__segment_derivatives)
        return user_positions
    
user_details = GetUserDetails(growwapi=growwapi)
user_details.getUserHoldings()
```

# Trading Strategy

```python
class TradingStrategy:
    def __init__(self, name):
        self.__name = name
    
    def generateSignal(self, price):
        print("will be over_ridden")
        return "Hold"
    
    @property
    def name(self):
        return self.__name
```

```python



class SmaTradingStrategy(TradingStrategy):
    def __init__(self, swindow, lwindow):
        super().__init__("SMA Strategy")
        self.__swindow = swindow
        self.__lwindow = lwindow
        
    def generateSignal(self, price_data: List[int]):
        if (len(price_data[-self.__lwindow:]) < self.__lwindow):
            return "Hold"
        
        short_avg = sum(price_data[-self.__swindow:]) / self.__swindow
        long_avg = sum(price_data[-self.__lwindow:]) / self.__lwindow
        
        if(short_avg > long_avg):
            return "Buy"
        elif(short_avg < long_avg):
            return "Sell"
        else:
            return "Hold"
        
    @property
    def swindow(self):
        return self.__swindow
    
    @property
    def lwindow(self):
        return self.__lwindow
```

```python


class TradeManager:
    def __init__(self, strategy: TradingStrategy, price_data: List[int]):
        self.__strategy = strategy
        self.__price_data = price_data
        self.__signal = self.__strategy.generateSignal([self.price_data])
        self.__time = ''
        self.__trades = []

    @property
    def strategy_name(self):
        return self.__strategy.name
    
    @property
    def price_data(self):
        return self.__price_data

    @property
    def signal(self):
        return self.__signal
    
    @property
    def time(self):
        self.__time = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        return self.__time
    
    @property
    def list_trades(self):
        return self.__trades

    def add_trade(self):
        time = datetime.datetime.now        
        trade = {
            'strategy_name': self.strategy_name,
            'price': self.price_data,
            'time': time,
            'signal': self.signal
        }
        self.list_trades.append(trade)
        print(f"Trade added: {trade}")

    def remove_trade(self, trade):
        if trade in self.list_trades:
            self.__trades.remove(trade)
            print(f"Trade removed: {trade}")
        else:
            print("Trade not found.")

    def execute_trade(self):
        signal = self.__strategy.generateSignal([self.price_data])
        print(f"Executing trade with signal: {signal}")
        self.add_trade()
```

```python
sma_strategy = SmaTradingStrategy(swindow=3, lwindow=6)
price_data = [1, 2, 3, 4, 5, 6, 7, 8, 9]
trade_manager = TradeManager(strategy=sma_strategy, price_data=price_data)
```

```python
trade_manager.execute_trade()
```
